```python
import numpy as np 

import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

!conda install -c conda-forge geopy --yes 
from geopy.geocoders import Nominatim 

import requests 
from pandas.io.json import json_normalize 


import matplotlib.cm as cm
import matplotlib.colors as colors


from sklearn.cluster import KMeans


from bs4 import BeautifulSoup

import xml

!conda install -c conda-forge folium=0.5.0 --yes
import folium # map rendering library

print('Libraries imported.') #all the libraries we need have been loaded
```

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(url,'lxml') #Wikipedia page scrapping
```

```python
table_post = soup.find('table')
fields = table_post.find_all('td')

postcode = []
borough = []
neighbourhood = []

for i in range(0, len(fields), 3):
    postcode.append(fields[i].text.strip())
    borough.append(fields[i+1].text.strip())
    neighbourhood.append(fields[i+2].text.strip())
        
df_postc = pd.DataFrame(data=[postcode, borough, neighbourhood]).transpose()
df_postc.columns = ['Postcode', 'Borough', 'Neighbourhood']
df_postc.head() #finding postal code by the Borough and Neighborhood
```

```python
df_postc['Borough'].replace('Not assigned', np.nan, inplace=True)
df_postc.dropna(subset=['Borough'], inplace=True) # removing NA and then compiling the table

df_postc.head()
```

```python
df_postcode = df_postc.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df_postcode.columns = ['Postcode', 'Borough', 'Neighbourhood']
df_postcode
```

```python
df_postcode['Neighbourhood'].replace('Not assigned', "Queen's Park", inplace=True)

df_postcode
```

```python
df_postcode.shape
```

```python
df_geos = pd.read_csv('http://cocl.us/Geospatial_data')
df_geos.columns = ['Postcode', 'Latitude', 'Longitude']
```

```python
df_postc = pd.merge(df_postcode, df_geos, on=['Postcode'], how='inner')

df_tor = df_postc[['Borough', 'Neighbourhood', 'Postcode', 'Latitude', 'Longitude']].copy()

df_tor.head()
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of the City of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_tor['Latitude'], df_tor['Longitude'], df_tor['Borough'], df_tor['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=3,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3199cc',
        fill_opacity=0.3,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
